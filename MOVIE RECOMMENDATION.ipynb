```python
columns_name=['user_id','item_id','rating','timestamp']
df=pd.read_csv('u.data',sep="\t",names=columns_name)
df.head(10)
```

```python
movies_title=pd.read_csv('u.item',sep="\|",header=None)
```

```python
movies_title.shape
```

```python
movies_title.head()
```

```python
movies_titles=movies_title[[0,1]]
```

```python
movies_titles.columns=["item_id","title"]
```

```python
movies_titles.head()
```

```python
df=pd.merge(df,movies_titles,on="item_id")
```

```python
df.head()
```

```python
ratings=pd.DataFrame(df.groupby('title').mean()['rating'])
```

```python
ratings.head()
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title').count()['rating'])
```

```python
movie_mat=df.pivot_table(index="user_id",columns="title",values="rating")
```

```python
movie_mat.head()
```

```python
AngryMen_user_ratings=movie_mat['12 Angry Men (1957)']
similar_to_AngryMen=movie_mat.corrwith(AngryMen_user_ratings)
corr_Angry_Men=pd.DataFrame(similar_to_AngryMen,columns=['correlation'])
corr_Angry_Men.dropna(inplace=True)
corr_Angry_Men.sort_values('correlation',ascending=False).head(10)
corr_Angry_Men=corr_Angry_Men.join(ratings['num of ratings'])
corr_Angry_Men.head()

corr_predict=corr_Angry_Men[corr_Angry_Men['num of ratings']>100].sort_values('correlation',ascending=False)
corr_predict.head()
```

```python
def predict_movies(movie_name):
    movie_user_ratings=movie_mat[movie_name]
    similar_to_movie=movie_mat.corrwith(movie_user_ratings)
    corr_movie=pd.DataFrame(similar_to_movie,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    corr_movie= corr_movie.join(ratings['num of ratings'])
    predictions=corr_movie[corr_movie['num of ratings']>100].sort_values('correlation',ascending=False)
    return predictions
```

```python
moviename=input("Enter a movie name ")
predict_my_movie=predict_movies(moviename)
print("\n\n...............Predicted top 5 movies based on the searched movie name............")
predict_my_movie.head(6)
```
